```python
import pandas as pd
df = pd.read_csv("nbaHomeWinLossModelDataset.csv")
print(df)
```

```python
df.head(10)
```

```python
df = df.dropna()
print(df)
```

```python
df[df.isnull().any(axis=1)]
df
```

```python
train = df[df["SEASON"] == ("2020-21" or "2021-22")].copy()
test = df[df["SEASON"] == "2022-23"].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ["HOME_LAST_GAME_OE", "HOME_LAST_GAME_HOME_WIN_PCTG", "HOME_NUM_REST_DAYS"]
target = "HOME_W"
```

```python
reg.fit(train[predictors], train[target])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test[target], test["predictions"])
```
